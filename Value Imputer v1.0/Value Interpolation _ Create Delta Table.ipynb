```python
df = spark.read.format("csv").option("header","true").load("<name>.csv")
# df now is a Spark DataFrame containing CSV data from "Files/Dataset_Value_Interpolation.csv".
display(df)
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
from pyspark.sql.types import IntegerType, BooleanType, DateType

# change column type
df_new = df \
    .withColumn("ID", col("ID").cast(IntegerType())) \
    .withColumn("Value", col("Value").cast(IntegerType())) \
    .withColumn("Predict", col("Predict").cast(IntegerType())) \
    #.withColumn("Dataset_Include", col("Dataset_Include").cast(IntegerType()))

# print schema
df_new.printSchema()
```

```python
from delta.tables import *

DeltaTable.createIfNotExists(spark) \
    .tableName("value_interpolation_dataset_v3") \
    .addColumn("ID", "INT") \
    .addColumn("Name", "STRING") \
    .addColumn("Value", "INT") \
    .addColumn("Predict", "INT") \
    .execute()
    #.addColumn("Dataset_Include", "INT") \
```

```python
df_new.write.format("delta").mode("overwrite").saveAsTable("value_interpolation_dataset_v3")
```
