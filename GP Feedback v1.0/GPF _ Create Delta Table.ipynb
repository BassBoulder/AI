```python
df = spark.read.format("csv").option("header","true").load("Files/ONS_GP_Rating_Jan24_Feb24.csv")
# df now is a Spark DataFrame containing CSV data from "Files/Dataset_Value_Interpolation.csv".
display(df)
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
from pyspark.sql.types import IntegerType, BooleanType, DateType, StringType

# change column type
df_new = df \
    .withColumn("Age_Band", col("Age_Band").cast(StringType())) \
    .withColumn("Response", col("Response").cast(StringType())) \
    .withColumn("Happy", col("Happy").cast(StringType())) \
    #.withColumn("Dataset_Include", col("Dataset_Include").cast(IntegerType()))

# print schema
df_new.printSchema()
```

```python
from delta.tables import *

DeltaTable.createIfNotExists(spark) \
    .tableName("ONS_GP_Rating_Jan24_Feb24_Dataset_v1") \
    .addColumn("Age_Band", "STRING") \
    .addColumn("Response", "STRING") \
    .addColumn("Happy", "INT") \
    .execute()
    #.addColumn("Dataset_Include", "INT") \
```

```python
df_new.write.format("delta").mode("overwrite").option("mergeSchema", "true").saveAsTable("ONS_GP_Rating_Jan24_Feb24_Dataset_v1")
```
